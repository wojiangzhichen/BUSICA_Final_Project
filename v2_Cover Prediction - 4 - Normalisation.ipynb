## Normalisation

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        result[feature_name] = df[feature_name]/ max_value
    return result
```

The features csv be normalised by this stage, but the data size is too big, and the data should be dealed with again for simplification (only normalised by the maximum value).

```python
X_normalised = normalize(X)
```
