### Split train and test data

```python
from sklearn.cross_validation import train_test_split
```

```python
train_d, test_d, train_lab, test_lab = train_test_split(x, y)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Set this range for the logistic regression model
c_range = np.linspace(start=0.001,stop=5,num = 5) 
```

```python
for c in c_range:
    lr=LogisticRegression(C=c)
    lr.fit(train_d, train_lab) 
    print('C=',c,', Prediction accuracy =', lr.score(test_d, test_lab))
```

From the fifth notebook, we know that the accuracy goes up when c increases, but a large c may be very weird; the data distribution is similar, and the accuracy is highest around c=5 in the specified range. So the overall accuracy is 60.1%.

#### fit the whole dataset

```python
lr = LogisticRegression(C = 5)
lr.fit(x,y)
```

## Table of features randed by importance

```python
#================Plot==================
importances=pd.Series(lr.coef_[2],index=features)
importances.sort_values(inplace=True,ascending=False)
importances.plot(kind="bar", figsize=(10,10));
plt.show()
```

From this point on,we can conclude that features that carry significance are: 'total_gross_sales', 'total_tip', 'total_cost', 'total_discount', 'total_seats', 'check_time_difference_in_seconds', 'total_stock_cost'.
