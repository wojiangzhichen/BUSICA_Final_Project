```python
x = pd.read_csv('D:/Project_Restaurant_Chain/Full/features_normalised.csv')
```

```python
y = pd.read_csv('D:/Project_Restaurant_Chain/Full/total_covers.csv')
```

```python
x.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
y.drop(['Unnamed: 0'], axis=1, inplace=True)
```

## Supervised Models

Due to the previous results, features that carry significance are: 'total_gross_sales', 'total_tip', 'total_cost', 'total_discount', 'total_seats', 'check_time_difference_in_seconds', 'total_stock_cost', others will be discarded.

```python
remained_features = ['total_gross_sales', 'total_tip', 'total_cost', 'total_discount', 'total_seats',\
                     'check_time_difference_in_seconds', 'total_stock_cost']
```

```python
x_remained = x[remained_features]
```

### Split train and test data

```python
from sklearn.cross_validation import train_test_split
```

```python
train_d, test_d, train_lab, test_lab = train_test_split(x_remained, y)
```

```python
print(len(train_d))
print(len(test_d))
print(len(train_lab)) 
print(len(test_lab))
```

## XGBoost

```python
import xgboost as xgb
```

```python
# Use the parameters obtained last time
xgb_model = xgb.XGBClassifier(max_depth=10,
                              n_estimators=50,
                              learning_rate=0.1,
                              gamma=0)
xgb_model.fit(train_d, train_lab)
```

```python
xgb_model.score(test_d, test_lab)
```

### Now the prediction accuracy comes up to 72.8%.

```python
# feature importance
for feature, importance in zip (remained_features,xgb_model.feature_importances_):
    print(feature,':',importance)
```

## Features ranked by importance 

```python
importances=pd.Series(xgb_model.feature_importances_,index=x_remained.columns)
importances.sort_values(inplace=True,ascending=False)
importances.plot(kind="bar", figsize=(10,10));
plt.show()
```
