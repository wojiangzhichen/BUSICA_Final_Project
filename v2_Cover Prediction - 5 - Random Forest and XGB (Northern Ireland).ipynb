## Supervised Models by region (Northern Ireland)

Choose restaurants in Northern Ireland, which has a small size of data, to do the model.

```python
Area_columns = ['Area_East England', 'Area_East Midlands',\
                'Area_Greater London', 'Area_Home Counties', 'Area_Ireland',\
                'Area_North East', 'Area_North West', 'Area_Northern Ireland',\
                'Area_Scotland', 'Area_South East', 'Area_South West', 'Area_Wales',\
                'Area_West Midlands', 'Area_Yorkshire & Humberside']
```

```python
other_features =['total_gross_sales', 'total_tip', 'total_discount',\
                 'total_cost', 'total_stock_cost', 'total_seats', 'total_inside_seats',\
                 'ground_seats', '1st_floor_seats', 'bar_seats', 'outside_seats',\
                 'private_dining_covers', 'rating_A',\
                 'rating_B', 'rating_C', 'rating_D', 'rating_E', 'rating_F',\
                 'check_time_difference_in_seconds']
```

### Split train and test data

```python
from sklearn.cross_validation import train_test_split
```

```python
train_d_Northern_Ireland, test_d_Northern_Ireland, train_lab_Northern_Ireland, test_lab_Northern_Ireland = train_test_split(x_Northern_Ireland, y_Northern_Ireland)
```

```python
print(len(train_d_Northern_Ireland))
print(len(test_d_Northern_Ireland))
print(len(train_lab_Northern_Ireland)) 
print(len(test_lab_Northern_Ireland))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=500, max_features=2, oob_score=True)

# Train the model
rfc.fit(train_d_Northern_Ireland, train_lab_Northern_Ireland)
```

```python
rfc.score(test_d_Northern_Ireland, test_lab_Northern_Ireland)
```

```python
print ("OOB accuracy for chains is ", rfc.oob_score_)
```

```python
importances=pd.Series(rfc.feature_importances_,index=train_d_Northern_Ireland.columns)
importances.sort_values(inplace=True,ascending=False)
importances.plot(kind="bar", figsize=(10,10));
plt.show()
```

## XGBboost

```python
import xgboost as xgb
```

```python
max_depth = [5,10]
n_estimators = [50, 100]
learning_rate = np.linspace(0.01,0.7,10)
gamma = [0,1,0.5]
```

```python
# find the best parameters
best_score = 0
max_i = 0
max_j = 0
max_h = 0
max_k = 0
for i in max_depth:
    for j in n_estimators:
        for h in learning_rate:
            for k in gamma:
                xgb_model = xgb.XGBClassifier(max_depth=i,
                            n_estimators=j,
                            learning_rate=h,
                            gamma=k)
                xgb_model.fit(train_d_Northern_Ireland, train_lab_Northern_Ireland)
                if xgb_model.score(test_d_Northern_Ireland, test_lab_Northern_Ireland) > best_score:
                    best_score = xgb_model.score(test_d_Northern_Ireland, test_lab_Northern_Ireland)
                    max_i = i
                    max_j = j
                    max_h = h
                    max_k = k
print('Out of 160 possible combinations, the best prediction is:')
print('max_depth=',max_i,'n_estimators=',max_j,'learning_rate=',max_h,'gamma=',max_k)
print('The best prediction=',best_score)
```

Out of 160 possible combinations, the best prediction is:
max_depth= 10 n_estimators= 50 learning_rate= 0.0866666666667 gamma= 0.
The best prediction= 0.71911786.

```python
# Use the parameters above
xgb_model = xgb.XGBClassifier(max_depth=10,
                              n_estimators=50,
                              learning_rate=0.0866666666667,
                              gamma=0)
xgb_model.fit(x_Northern_Ireland, y_Northern_Ireland)
```

```python
# feature importance
for feature, importance in zip (other_features,xgb_model.feature_importances_):
    print(feature,':',importance)
```

```python
importances=pd.Series(xgb_model.feature_importances_,index=x_Northern_Ireland.columns)
importances.sort_values(inplace=True,ascending=False)
importances.plot(kind="bar", figsize=(10,10));
plt.show()
```
